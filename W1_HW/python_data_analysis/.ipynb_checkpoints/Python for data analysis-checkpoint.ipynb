#### Check if there are any extreme sepal length values (extreme value criterion: less or greater than 3.9 standard diviations away from the mean)

```python
mean_sl = df['sepal length (cm)'].mean()
std_sl = df['sepal length (cm)'].std()
df.loc[abs(df['sepal length (cm)']-mean_sl) > 3.9*std_sl , ['sepal length (cm)']]
```

```python
mask=np.greater(abs(df['sepal length (cm)']-mean_sl), 3.9*std_sl)
df[mask]['sepal length (cm)']
```

No extreme values for `sepal length`.

#### What about the other three features? Define a function that checks for extreme values for each of the features.

```python
def check_extreme(data, c = 3.9): # c is the extreme criterion   
    mean = data.mean()
    std = data.std()
    z=(data-mean)/std
    mask = np.greater(abs(z) , c) #standardized > 3.9 then True
    s=data[mask].dropna(how = 'all') #left rows w/ any values not NaN
    if s.shape[0] == 0:
        print("No extreme values.")
    else:
        print("Extreme Values Found!")
    return(s)   
check_extreme(df)
```

#### Optional: Usually boxplots (see below) treat as “outliers” those values that are outside of 1.5 times the interquartile range. You can try to define a function that checks for boxplot outliers too.

```python
def check_extreme_2(data, c = 1.5): # c is the extreme criterion   
    #mean = data.mean()
    IQR = data.describe().loc['75%']-data.describe().loc['25%']
    #print(IQR)
    high = data.describe().loc['75%']+ c*IQR
    low = data.describe().loc['25%']- c*IQR
    #print(high)
    #print(low)
    mask = np.logical_or(data>high, data<low) #>1.5 IQR then True
    return(data[mask].dropna(how = 'all'))

    
check_extreme_2(df)   
```
